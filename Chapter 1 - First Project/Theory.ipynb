## RMSE - Root Mean Square Error
RMSE are using for measure standard deviation of the errors the system makes in predictions. Calculation of RMSE corresponds to *Euclidian norm*.

**Where**: 
 + **n** - number of train examples used for measure RMSE error.
 + **i** - index of current iterration.
 + **predicted** - predicted value of **i**-sample
 + **actual** - actual value of **i**-sample
---

## MAE - Mean Absolute Error
Metric of measure model prediction errors, is good to use when train data contains many "noise" samples. Calculation of RMSE corresponds to *Manhattan norm*.

**Where**:
 + **N** - number of train examples used for measure RMSE error.
 + **i** - index of current iterration.
 + **y** - actual value of **i** sample.
 + **ŷ** - predicted value of **i** sample.
